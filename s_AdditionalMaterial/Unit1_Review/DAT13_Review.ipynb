## Python Control Flow
* If
* For
* While

```python
# If
numbers = range(1,31)

# Classic IF program... fizz buzz
# Print out each number in the range 'numbers'. But for multiples of 3, print "fizz". For multiples of 5, print "buzz". 
# For multiples of both 3 AND 5, print "fizzbuzz".

# Add to the code below to get fizzbuzz working.

for n in numbers:
    if n%3 == 0:
        print "fizz"
    elif n%5 == 0:
        print "buzz"
    else:
        print n
```

```python
# For
numbers = range(10)
for n in numbers:
    print n

# Usage Example: list comprehension
squares = [n*n for n in numbers]
print squares
```

```python
# While/Break

# Right now, Monrovia is reporting about 200 Ebola cases each week. 
# The city has a population of about 1 million people. So that means 1 out of every 5,000 people gets 
# infected with Ebola each week. That puts the risk of getting Ebola each week in Monrovia right about 
# on par with the risk of dying during childbirth in the U.S. (1 out 5,500 births) or about two times 
# the risk of dying in a traffic accident each year (1 in 9,100 people).

# Of course, the longer you stay in Monrovia, the more your risk goes up because you are continually in
# situations where you might be exposed to the virus. For instance, a full year in Monrovia raises your 
# chance of getting Ebola to around 1 percent — assuming the epidemic doesn't grow any faster than it is now.


# Assuming that 1% is constant chance of contracting Ebola every year you live in Monrovia.
# Obviously, this ignores compounding chances, but let's pretend...

import random
chance = 0.01

has_ebola = False
years = 0
while has_ebola == False:
    years += 1
    roll = random.random()
    # print "Year %d roll = %f.4" % (years, roll)
    if roll < chance:
        has_ebola = True
    if years > 100:
        break

print "Its been %d years and my ebola status is: %s" % (years, has_ebola)
```
