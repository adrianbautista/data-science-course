# Acquisition

```python
df = pd.read_table('./imdb_top_10000.txt', header=None)
```

# Cleaning

```python
column_names = ['ID1', 'title', 'year', 'rating', 'ID2', 'duration', 'genres']
df.columns = column_names
df.head()
```

```python
# data.duraction.apply(lambda y: y.split('(')[0]))

df['duration'] = df['duration'].str.replace(' mins.', '')
df.head()
```

```python
df['title'] = df['title'].str.replace('\s\(\d+\)', '')
df.head()
```

```python
df['year'] = pd.to_datetime(df['year'], format="%Y")
df.head()
```

```python
df.drop(['ID1', 'ID2'], axis=1, inplace=True)
df.head()
```

```python
df = df.dropna()
```

```python
genreset = set()

for x in df.genres:
    genreset.update(g for g in x.split('|'))
    
print genreset
```

```python
# for all the genres in the set, if the given movie contains the genre, return true for that genre
# a column is being created for each genre in genreset

for genre in genreset:
    df[genre] = [genre in movie.split('|') for movie in df.genres]
    
df.drop('genres', axis=1, inplace=True)
df.head()
```

```python
df.duration = df.duration.apply(lambda y: int(y))
df.head()
```

```python
df.describe()
```

# Explore

```python
df_by_year = df.groupby('year')
rating_year = df_by_year['rating'].aggregate([np.mean])
rating_year.head()
```
