### relationship between sleep_rem and brain_weight

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats

from pandas.tools.plotting import scatter_matrix
from matplotlib import pyplot as plt
```

```python
# clone dataset and clean its data by dropping rows with NA values

spat_cleaned = pd.DataFrame(spat)
spat_cleaned.brainwt.dropna(inplace=True)
spat_cleaned.sleep_rem.dropna(inplace=True)

# add two columns of log-transformed data of brain weight and sleep rem

spat_cleaned['log_brainwt'] = np.log(spat_cleaned.brainwt)
spat_cleaned['log_sleep_rem'] = np.log(spat_cleaned.sleep_rem)

# create a general figure with 2 subplots, fig is the figure and axes is the list of subplots

fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Perform OLS analysis of brainwt's normal relationship to sleep rem 

normal_model = smf.ols(formula='sleep_rem ~ brainwt', data=spat_cleaned)
results = normal_model.fit()

print 'NORMAL FIT SUMMARY'
print(results.summary())
```

```python
# Perform OLS analysis of brainwt's logarithmic relationship to sleep rem 

log_model = smf.ols(formula='log_sleep_rem ~ log_brainwt', data=spat_cleaned)
log_results = log_model.fit()
print 'LOG FIT SUMMARY'
print(log_results.summary())
```
