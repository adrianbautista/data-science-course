```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier

%pylab inline
```

```python
spectra_data_set = pd.read_csv('micromass/pure_spectra_matrix.csv', sep=';', header=None)
spectra_data_set.shape
```

```python
serial_index = pd.read_csv('micromass/pure_spectra_metadata.csv', sep=';', header=0)
serial_index.head()
```

```python
serial_index['bacteria_genera'] = serial_index['Species'].apply(lambda x: x.split('.')[0])
```

```python
spectra_data_set.set_index(serial_index['bacteria_genera'], inplace=True)
spectra_data_set.head()
```

```python
X = pd.read_csv('micromass/pure_spectra_matrix.csv', sep=';', header=None)

# A bit sparse
print 'fraction non-zero:', (X!=0).sum().sum() / np.prod(X.shape)

Y = pd.read_csv('micromass/pure_spectra_metadata.csv', sep=';')

Y['gram'] = map(lambda x: x.split('.')[0] in
                          ['QWP', 'VVJ', 'QBG', 'RTO'],
                Y.Species)

# About 31% gram positive
print 'fraction gram positive:', np.mean(Y.gram)

tree = DecisionTreeClassifier(max_depth=1)
tree.fit(X, Y.gram)
print 'stump accuracy on all data:', tree.score(X, Y.gram)

pca = PCA(n_components=1)
X1 = pca.fit_transform(X)
tree.fit(X1, Y.gram)
print 'stump accuracy on one principal component', tree.score(X1, Y.gram)
```
