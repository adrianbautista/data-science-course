# Looking at real-time wholesale electricity prices in NYS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix

%pylab inline
```

#Acquire

The data was acquired at the New York Independent System Operator (NYISO) website (http://www.nyiso.com/public/markets_operations/market_data/pricing_data/index.jsp). The NYISO is responsible for operating the wholesale electricity market in New York State.

The dataset of interest is the real-time wholesale prices for the 15 different submarkets in New York State for September 25th. The prices are given for every 5-minute interval from 12:05 am to 6:15 pm.

```python
data = pd.read_csv('real_time_energy_pricing_nys.csv') #reading from the csv file

data.head() #Let's have a look at the first five rows of the dataset to get a feel for the data
```

We  define the columns of the dataset as follows:

Time Stamp
Name: name of the submarket
PTID: submarket ID number

LBMP ($/MWHr): Locational Based Marginal Price (LBMP)is the cost of energy production plus the transportation (losses and congestion) cost in dollars per MWHr

Marginal Cost Losses ($/MWHr): The marginal losses component of the LBMP at a bus (terminal in transmission lines) is positive when a small increase in generation at the bus causes an overall decrease in transmission losses. Conversely, the marginal losses component of the LBMP at a bus is negative when a small increase in generation at the bus causes an overall increase in transmission losses.

```python
submarket = data.groupby(['name'])
submarket_sorted = submarket.aggregate(np.mean)

Sorted = submarket_sorted.sort(['lbmp'], ascending=False)

Sorted.head(15)
```

Not very interesting...
Let's look at histograms and scatterplots for the three columns of interest: lbmp, marginal losses and marginal congestion:

```python
pd.scatter_matrix(submarket_sorted[['lbmp','marginal_losses','marginal_congestion']])
submarket_sorted[['lbmp','marginal_losses','marginal_congestion']].hist()
```

```python
print scatter_matrix(submarket_sorted, figsize=[10, 10], alpha=0.2, diagonal='kde')
```

Normal Distribution?

```python
norm = pd.DataFrame({'d': np.random.normal(size=60)})
n_bins = np.abs(submarket_sorted['lbmp'].max() - submarket_sorted['lbmp'].min())
fig, axes = plt.subplots(nrows=2, ncols=2,)

norm['d'].plot(ax=axes[0, 0], kind='kde')
norm['d'].hist(ax=axes[0, 1], bins=n_bins/2)
submarket_sorted['lbmp'].plot(ax=axes[1, 0], kind='kde')
submarket_sorted['lbmp'].hist(ax=axes[1, 1], bins=n_bins/2)

print fig
```

Doesn't really look like a normal distribution

```python
#Let's confirm this with a qq plot

qq = sorted(np.random.normal(submarket_sorted['lbmp'].mean(), submarket_sorted['lbmp'].std(), len(submarket_sorted)))
lbmp = sorted(submarket_sorted['lbmp']) #need to make sure the data is sorted, since random gives us data unordered, name is confusing here

plt.plot(qq, lbmp, 'bo')
plt.show()
```

This qq plot confirms that the lbmp data doesn't have/follow a normal distribution.
What about marginal_losses?

```python
norm = pd.DataFrame({'d': np.random.normal(size=60)})
n_bins = np.abs(submarket_sorted['marginal_losses'].max() - submarket_sorted['marginal_losses'].min())
fig, axes = plt.subplots(nrows=2, ncols=2,)

norm['d'].plot(ax=axes[0, 0], kind='kde')
norm['d'].hist(ax=axes[0, 1], bins=n_bins/2)
submarket_sorted['marginal_losses'].plot(ax=axes[1, 0], kind='kde')
submarket_sorted['marginal_losses'].hist(ax=axes[1, 1], bins=n_bins/2)

print fig
```

```python
import statsmodels.api as sm
import scipy.stats as stats

# Check sm.qqplot() configurations with shift+tab to see how you can change the distribution you are testing against!
fig = sm.qqplot(submarket_sorted['marginal_losses'], dist=stats.distributions.norm, line='s')
plt.show()
```

Clearly this isn't a normal distribution either

# Summary

Finally, let's see if there is any correlation between marginal losses and lbmp

```python
import statsmodels.formula.api as smf
# OLS, or ordinary least squares, takes a y (dependent variable) and X (independent variables) (formula = y ~ X)
# Below, we copy the data frame and remove the na variables, and create a single variable linear model
# to return a test statistic and p-value, to see how strong of a relationship bodyweight and brainweight have.

sub_cleaned_up = pd.DataFrame(submarket_sorted) #creating a copy
sub_cleaned_up['log_lbmp'] = np.log(sub_cleaned_up['lbmp'])
sub_cleaned_up['log_marginal_losses'] = np.log(sub_cleaned_up['marginal_losses'])

fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(sub_cleaned_up.lbmp, sub_cleaned_up.marginal_losses, 'go')

model = smf.ols(formula='lbmp ~ marginal_losses', data=sub_cleaned_up) #simplest linear regression you can run, we want to understand the relationship between brain size and body weight, see if they correlate
results = model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())
print

axes[1].plot(sub_cleaned_up.log_lbmp, sub_cleaned_up.log_marginal_losses, 'mo')

log_model = smf.ols(formula='log_lbmp ~ log_marginal_losses', data=sub_cleaned_up)
log_results = log_model.fit()
print 'LOG-LOG FIT SUMMARY'
print(log_results.summary())

print fig
```

Both R squared values are very low: lbmp and marginal cost losses are not correlated.
