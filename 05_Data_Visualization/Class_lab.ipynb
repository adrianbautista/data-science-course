```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
# print X
```

```python
plt.plot(X, C)
plt.plot(X, S)
```

```python
plt.figure(figsize=(8, 6)) # overall size of canvas

plt.subplot(1, 1, 1) # number of plots inside the canvas (number of rows, number of columns, where you're plotting)

plt.plot(X, C) # plot the data

plt.xlim(-4.0, 4.0) # range of the x-axis and its ticks
```

```python
plt.figure(figsize=(8, 6))
plt.xlim(-4.0, 4.0)

plt.subplot(2, 2, 1)
plt.plot(X, C)

plt.subplot(2, 2, 2)
plt.plot(X, S)

plt.subplot(2, 2, 3, axisbg='gray')
plt.plot(X, C)

plt.subplot(2, 2, 4, axisbg='gray')
plt.plot(X, S)

# plt.subplot(2, 2, 4) => OUT OF RANGE ERROR
```
