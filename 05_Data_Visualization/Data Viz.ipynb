```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(10000), 'k',label='one')
ax.plot(np.random.randn(10000).cumsum(), 'k--',label='two')
ax.plot(np.random.randn(10000).cumsum(), 'k,',label='three')
ax.set_title('Three Lines')

ax.legend()
```
